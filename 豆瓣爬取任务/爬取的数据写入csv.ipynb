```python
import csv
from lxml import etree
import requests

url = 'https://movie.douban.com/top250'

head = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36 Edg/130.0.0.0'
}

urls = ['https://movie.douban.com/top250?start={}&filter='.format(str(i * 25))for i in range(10)]
count = 1

res = requests.get(url,headers=head)
res.encoding = res.apparent_encoding
html = res.text
# print(res.status_code)
# print(html)
s = etree.HTML(html)
li = s.xpath('//*[@id="content"]/div/div[1]/ol/li')
src1 = s.xpath('/html/body/div[3]/div[1]/div/div[1]/ol/li')
ping_1 = s.xpath('//*[@id="content"]/div/div[1]/ol/li')
ping_2 = s.xpath('//*[@id="content"]/div/div[1]/ol/li')
director = s.xpath('//*[@id="content"]/div/div[1]/ol/li')


for url in urls:
    res = requests.get(url=url,headers=head)
    html = etree.HTML(res.text)
    li = html.xpath('//*[@id="content"]/div/div[1]/ol/li')
    
    num = 1
    with open('./douban_data.csv','a+',newline='',encoding='utf-8')as f:
        for i in li:
            
            rank = i.xpath('./div/div[2]/div[1]/a/span[1]/text()')[0]       #排名
            src = i.xpath('./div/div[2]/div[1]/a/@href')[0]                 #链接
            score = i.xpath('./div/div[2]/div[2]/div/span[2]/text()')[0]    #评分
            riew = i.xpath('./div/div[2]/div[2]/div/span[4]/text()')[0]     #评论人数
            direct = i.xpath('./div/div[2]/div[2]/p[1]/text()')[0].strip()  #导演

            item=[rank,src,score,riew,direct]
            f_csv = csv.writer(f)
            f_csv.writerow(item)

            print(rank,src,score,riew,direct)
            count+=1
```
