```python
import pandas 
import matplotlib.pyplot as plt
# 读取 csv
data1 = pandas.read_csv('C:\\Users\\26440\\Desktop\\jupyter_ipynb\\豆瓣爬取任务\\豆瓣电影Top250.csv',encoding='utf-8',nrows=10)
# print(data1)

#图示1 plt.plot 折线图
plt.rcParams['font.sans-serif'] = ['SimHei']    #解决图画中文乱码方法
plt.figure(figsize=(10,10),dpi=360)         #为视图修改大小,dpi 表示分辨率
plt.plot(data1['电影中文名'],data1['评分'])

plt.title('豆瓣Top10电影排行')      # 图表标题
plt.xlabel('电影名称')              # X横轴
#plt.xticks([0,2,4,6,8,10])          #规定x轴刻度
plt.ylabel('电影评分')              # Y纵轴
#plt.yticks([0,2,4,6,8,10])          #规定y轴刻度
plt.show()
```

```python
#图示2  plt.scatter 散点图
'''
    scatter 函数介绍
    1.marker 表示 绘制点的类型
    2.c 表示 点的颜色，若为一维数组则表示每个点的颜色，默认None
    3.x，y 表示x轴和y轴对应的数据
    4.s 表示点的大小
    5.alpha 表示点的透明度 接收float 
'''
plt.figure(figsize=(9,9),dpi=360)
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
plt.scatter(data1['电影中文名'],data1['评分'],marker='o',c='b')#散点图

plt.title('豆瓣前10评分')
plt.xlabel('电影名称')
plt.ylabel('电影评分')
plt.ylim(9,10)     #设置y轴范围

plt.show()
```
