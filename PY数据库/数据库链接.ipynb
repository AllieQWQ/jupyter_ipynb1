```python
import pymysql

# 1、创建数据库链接对象connect
db = pymysql.connect(
    host='localhost',
    port=3306,
    user='root',
    passwd='1024',  # 确保这是正确的密码
    db='student',  # 连接到student数据库
    charset='utf8'
)

# 2、获取游标对象 cursor
cursor = db.cursor()

try:
    # 3、执行SQL语句（核心步骤）
    # 修改为查询当前数据库中的所有表
    sql = 'SHOW TABLES;'
    cursor.execute(sql)
    
    # 获取查询结果
    tables = cursor.fetchall()
    
    # 打印所有表的名称
    for table in tables:
        print(f"Table: {table[0]}")
        
    # 如果你想查看每个表的内容，可以继续执行以下代码
    for table in tables:
        # 查询每个表的所有数据
        query_sql = f'SELECT * FROM {table[0]};'
        cursor.execute(query_sql)
        rows = cursor.fetchall()
        
        # 打印表的内容
        print(f"Contents of {table[0]}:")
        for row in rows:
            print(row)

except pymysql.MySQLError as e:
    print(f"An error occurred: {e}")

finally:
    # 4、关闭游标和数据库连接
    cursor.close()
    db.close()
```

```python
import pymysql

try:
    # 1、创建数据库链接对象connect
    db = pymysql.connect(
        host='localhost',
        port=3306,
        user='root',
        passwd='1024',
        db='student',
        charset='utf8'
    )

    # 2、获取游标对象 cursor
    cursor = db.cursor()

    # 3、执行SQL语句（核心步骤：建表、增删改查）
    sql = 'create table if not exists studentinfo(id varchar(100) not null primary key,name varchar(100) not null,age int not null);'
    cursor.execute(sql)

    '''
        因为插入的内容很多，但是格式单一，所以用占位符 '%s' 来代替
        重复多行插入数据，解决主键冲突问题 并检查是否有重复数据
    '''
    id = '5'
    name = 'Alen'
    age = 24

    # 检查是否存在相同ID的记录
    check_sql = 'select * from studentinfo where id = %s'
    cursor.execute(check_sql, id)
    result = cursor.fetchone()

    if result:
        print('该ID已存在，跳过插入操作。')
    else:
        # 插入数据
        sql1 = 'insert into studentinfo(id, name, age) values(%s, %s, %s);'
        cursor.execute(sql1, (id, name, age))
        print('插入数据成功！')

    # 更新数据的SQL命令
    update_id = '5'  # 假设要更新的ID为5
    new_name = 'Alice'
    new_age = 25
    update_sql = 'update studentinfo set name = %s, age = %s where id = %s'
    cursor.execute(update_sql, (new_name, new_age, update_id))
    print(f'更新ID为 {update_id} 的数据成功！')

    # 删除数据的SQL命令
    delete_id = '5'  # 假设要删除的ID为5
    del_sql = 'delete from studentinfo where id = %s'
    cursor.execute(del_sql, delete_id)
    print(f'删除ID为 {delete_id} 的数据成功！')

    # 提交事务 commit (只针对对数据库有修改的一些操作 -> 增删改查等操作 无需提交事务) rollback() 回滚当前事务
    db.commit()

except Exception as error:
    db.rollback()
    print('操作失败！', error)

finally:
    # 4、提交事务（只针对数据库有修改的一些操作，增删改查等操作不需要提交事务 ，关闭链接注意顺序）
    cursor.close()
    db.close()
```
