```python
import  requests
import  json
import  csv

url='https://ugc.dujia.qunar.com/c2b/list?limit=12&page=1&arrs='
r=requests.get(url)
#print(type(r.text))


#2解析json数据（提取字典关键信息）
 #json字符串
data = json.loads(r.text)
new = json.dumps(data)#son格式字符串
print(new)

#提取字典关键信息（地名，count）
sites=data['data']
for i in sites:
    print(i['name'],i['count'])

# 3. 提取关键信息（地名，count）
sites = data.get('data', [])
if not sites:
    print("未找到数据")
    exit()

```
