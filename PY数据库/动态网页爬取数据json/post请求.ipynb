```python
'''
json参数用于发送json格式数据 即'application/json' 
也有可能是表单数据 
'''
import requests
import json
import csv

def get_one_page(url):
    url = 'https://www.gov.cn/zhengce/qiye/trssearch/search/all'
    header = {
        'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36 Edg/135.0.0.0',
        'cookie':'wdcid=501e30dfe02ba0c6; SERVERID=dfa8896f873ec6687e25200a445d55c6|1744598269|1744598269; wdlast=1744598266; wdses=0217bd455e796c72; arialoadData=false; ariauseGraymode=false',
        'content-type':'application/json'       #post请求 参数为 json 格式
    }
    data = {"dbname":"","currpage":1,"pagesize":10,"searchWord":"","searchField":"title","searchSort":"time","DocType":"","publishDept":"","hqType":"","enterpriseType":"","enterpriseScale":"","lifeCycle":0}

    r = requests.post(url,headers = header,json = data)
    #print(r.json())
    return r.text

def get_parse_page(json_data):
    tilte_data = []
    #将json 数据（字符串） -> 字典
    one_theme_data = json.loads(json_data)
    two_theme_data = one_theme_data["data"]["list"]
    
    #逐行遍历每一篇文章
    for i in two_theme_data:
        title = i["title"]
        timestamp = i["publishDocTime"]
      
        tilte_data.append([title,timestamp])
        print(title,timestamp)
    return  tilte_data

def write_one_page(content,path):
    with open(path,'a',newline='',encoding='utf-8-sig') as f:
        writer = csv.writer(f)
        if f.tell() == 0:
            writer.writerow(['标题','时间戳'])
        writer.writerows(content)

if __name__ == '__main__':
    url = 'https://www.gov.cn/zhengce/qiye/trssearch/search/all'
    json_data = get_one_page(url)
    
    if json_data:
        content = get_parse_page(json_data)
        if content:
            write_one_page(content,'post实战.csv')
        else:
            print('解析数据失败！')
    else:
        print('获取数据失败')
    
```
