```python
'''
    2. 重复值处理，基于上述数据df，请完成以下操作：
        1）查找重复值：将全部重复值所在的行筛选出来;
        2) 查找指定重复值：查找片名列全部重复值；
        3) 删除全部的重复值;
        4) 删除全部的重复值，但保留最后一次出现的值

'''
# 1）查找重复值：将全部重复值所在的行筛选出来;
df.loc[df.duplicated(),:]

# 2) 查找指定重复值：查找片名列全部重复值；
df.loc[df.片名.duplicated(),:]

# 3) 删除全部的重复值;
df.drop_duplicates(inplace=True)    #之前数据是262rows 删除后只有252rows

# 4) 删除全部的重复值，但保留最后一次出现的值
df.drop_duplicates(keep='last')
```

```python
'''
    二、检测与处理缺失值
    3. 缺失值处理，基于上述数据df，请完成以下操作：
        1）查看具体每列有多少缺失值;
        2) 为了后面更方便的处理缺失值，查看全部缺失值所在的行；
        3）将缺失值出现的行全部删掉；
        4）将全部缺失值替换为 * ;
        5) 将评分列的缺失值，替换为上一个电影的评分;
        6) 将评价人数列的缺失值，用整列的均值进行填充;
'''
# 1）查看具体每列有多少缺失值;
df.isnull().sum()

# 2) 为了后面更方便的处理缺失值，查看全部缺失值所在的行；
df.loc[df.isnull().any(axis=1),:]

# 3）将缺失值出现的行全部删掉；
df.dropna(axis=1,how='any')

# 4）将全部缺失值替换为 * ;
df1 = df.fillna(value= '*')
df1

# 5) 将评分列的缺失值，替换为上一个电影的评分;
#df.loc[df.评分.isnull()==True]
df2 = df.评分.fillna(method='ffill')
df2

# 6) 将评价人数列的缺失值，用整列的均值进行填充; 直接在前面 df.评价人数.fillna()
df3 = df.评价人数.fillna(df.评价人数.mean())
df3
```

```python
'''
    三、检测与处理异常值
        4. 基于上述数据df，利用3∂原则根据df中的评分列过滤异常值
        5. 基于上述数据df，使用箱线图分析方法来识别评分列数据中的异常值
        绘制箱线图不能有空缺值
'''
import matplotlib.pyplot as plt 
df.loc[numpy.abs(df.评分-df.评分.mean()) > 3 * df.评分.std()]

df = df.dropna()
plt.figure(figsize=(10,8),dpi=720)
p = plt.boxplot(df.评分)    #画出箱线图
outlier = p['fliers'][0].get_ydata()
print(outlier)
plt.show()
```

```python
#以评分为例 用箱线图方法查看异常值
Q1 = df.评分.quantile(0.25)  #1/4分位 
Q3 = df.评分.quantile(0.75)  #3/4分位
IQR = Q3 - Q1   #四分位间距

low = Q1 - 1.5 * IQR    #下边界
high = Q3 + 1.5 *IQR    #上边界

#筛选评分异常值
df.loc[(df.评分 > high)|(df.评分 < low)]
```
