```python
# sort和argsort排序(np.sort 和 np.argsort)
import numpy as np 
np.random.seed(123)     #设置随机种子
arr1 = np.random.randint(1,10,size=(3,3))  #生成随机数组
print('创建的数组为：\n',arr1)

arr1.sort(axis=0)  #默认沿着横轴进行排序(axis = 0 代表纵轴;axis = 1 代表横轴)
print('纵轴数组为：\n',arr1)
arr1.sort(axis=1)
print('横轴数组为：\n',arr1)
```

```python
#间接排序：argsort函数（得到一个由整数构成的索引数组，索引值都是在新的序列中的位置）
np.random.seed(12)     #设置随机种子
arr2 = np.random.randint(1,10,size=10)  #生成随机数组
print('创建的数组为：\n',arr2)
r = arr2.argsort()  #返回调用后的结果
print(r)
```

```python
#lexsort 函数 可以一次性对满足多个键值的数组执行间接排序
a = np.array([3,2,6,4,5])
b = np.array([50,30,40,20,10])
c = np.array([400,300,600,100,200])
r = np.lexsort((a,b,c)) #多个键值排序是按照最后一个传入数据计算的(如果b，c的后俩数据相同，则比较a的后两位下标)
print(r) 

print('排序后的数组为：',list(zip(a[r],b[r],c[r]))) #r代表下标位置
```

```python
arr3 = np.random.seed(42)
arr3 = np.random.randint(1,10,size=10)  #一个包含重复元素的一维数组
print(arr3)
arr4 = np.sort(arr3)
print('重新排序后的数组:',arr4)
arr5 = np.argsort(arr3)
print(arr5)
```

```python
#去重与重复数据(例题)   unique;repeat;tile
'''
    tile 函数 和 repeat 函数 实现数据重复
    
'''
list1 = np.arange(5)
print('创建的数组为：\n',list1)
print('重复后的数组为：\n',np.tile(list1,3))
```

```python
#去重与重复数据
np.random.seed(123)
ar1 = np.random.randint(1,16,size=(3,3))
print('创建的3*3数组为：\n',ar1)
ar2 = np.repeat(ar1,2,axis=0)   #按照行 去重复
ar3 = np.unique(ar2,axis=0)            #对该数组的重复进行去除
ar4 = np.repeat(ar1,2,axis=1)   #按照列 去重复
print('按照行重复后的结果：\n',ar2)
print('去除重复后的结果：\n',ar3)
print('按照列重复后的结果：\n',ar4)

ar5 = np.random.randint(0,2,size=(2,2))
print('创建的8*8棋盘样式为：\n',np.tile(ar5,(4,4)))
```

```python
#教师案例
import numpy as np
arr = np.arange(20).reshape(4,5)
print('创建的数组：\n',arr)
arr.sum(axis=1)
arr.mean(axis=0)
arr.std()
arr.var()
arr.min(axis=1)
arr.max(axis=0)

print(np.argmin(arr,axis=1))    #返回数组最小元素的索引 np.argmin(arr)/arr.argmin()
print(np.argmax(arr))           #返回数组最大元素的索引
```

```python
#教师案例
arr9 = np.arange(1,4)
print('创建的数组：\n',arr9)
print('数组元素的累计求和为：',np.cumsum(arr9))     #计算所有元素的累计和
print('数组元素的累计积为：',np.cumprod(arr9))      #计算所有元素的累计积
```

```python
# 5. 统计分析
np.random.seed(43)
list2 = np.random.randint(1,100,size=(5,5))
print('生成的数组为：\n',list2)
print('数组的和为：',np.sum(list2))
print('数组的均值：',np.mean(list2))
print('数组的标准差：%.2f'%np.std(list2))
print('数组的方差：%.2f'%np.var(list2))

print('数组沿纵轴（列）的总和：',list2.sum(axis=0))
print('数组沿横轴（行）的总和：',list2.mean(axis=1))
print('数组的最大值：',np.max(list2),'数组的最小值：',np.min(list2))
print('数组减去数组中每一行的平均值:\n',list2 - np.mean(list2))
```
