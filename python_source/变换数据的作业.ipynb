```python
'''
    (1)读取来聘人员信息表(hr_job.csv)数据
    (2)将类别型数据中的缺失值填补为“未知”，将数值型缺失值填补为其对应特征的均值。
    (3)对所有的分类数据进行哑变量处理。
'''
import pandas
df = pandas.read_csv(
    'C:\\Users\\26440\\Desktop\\jupyter_ipynb\\python_source\\data\\hr_job.csv'
    ,encoding='gbk')

#(2)将类别型数据中的缺失值填补为“未知”，将数值型缺失值填补为其对应特征的均值。
df1 = df.select_dtypes(include=['object']).columns
df[df1] = df[df1].fillna('未知')
df

df2 = df.select_dtypes(include=['number']).columns
for i in df2:
    avg = df[i].mean()
    df[i].fillna(avg,inplace=True)
df[i]
df3 = pandas.get_dummies(df)
df3
```
