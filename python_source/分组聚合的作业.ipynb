```python
'''
    1）计算各区(district)的薪资(salary)均值;
    2) 重新按照上一题要求进行分组，但不使用 district 做为索引
    3）计算并提取平均薪资最高的区；
    4）计算不同行政区(district)，不同规模公司(companySize)出现的次数
    5）计算上一题，每个区出现的公司数量
    6）将数据按照 district、salary 进行分组，并查看各分组内容
    7）将数据按照 district、salary 进行分组，并查看西湖区薪资为 30000 的工作
'''
import pandas 
import numpy
df = pandas.read_csv(
    'C:\\Users\\26440\\Desktop\\jupyter_ipynb\\python_source\\data\\某招聘网站数据.csv'
    ,encoding='utf-8'
    )
df
```

```python
print('各区的薪资均值为：\n',df.groupby(by='district')['salary'].mean().
      apply(lambda x:round(x,2)))
print(df.groupby(by='district',as_index=False)['salary'])
#ascending 是 升序  ascending=False 则是降序
df.groupby(by='district',as_index=False)['salary'].mean().sort_values('salary',ascending=False).head(1)

#方法2：找出最大值
#df.groupby(by='district',as_index=False)['salary'].mean().sort_values('salary').max()
```

```python
#计算上一题，每个区出现的公司数量
print('不同规模公司出现的数量：\n',df.groupby(by='district')['companySize'].count())

#计算不同行政区(district)，不同规模公司(companySize)出现的次数
df.groupby(by=['district','companySize'])['companySize'].size()
```

```python
#数据按照 district、salary 进行分组，并查看各分组内容
data = df.groupby(by=['district','salary'])
data.groups
```

```python
group = data.get_group(('西湖区',30000))
group
```

```python
'''
    2.分组计算不同行政区(district)，薪水(salary)的最小值、最大值和平均值
'''
df.groupby(by='district')['salary'].agg([numpy.min,numpy.max,numpy.mean])
```

```python
#对不同岗位(positionName)进行分组，并统计其薪水(salary)中位数和得分(score)均值
df.groupby(by='positionName').agg({'salary':numpy.median,'score':numpy.mean})
```

```python
# 对不同行政区(district)进行分组，并统计薪水(salary)的均值、中位数、方差，以及得分(score)的均值
df.groupby(by='district').agg({'salary':[numpy.mean,numpy.median,numpy.var],'score':numpy.mean})
```

```python
#计算各行政区的企业领域（industryField）包含电商的总数（通过匿名函数）
df['Sum'] = df['industryField'].apply(lambda x: '电商' in x)
ecommerce_counts = df.groupby('district')['Sum'].agg(['sum'])
print(ecommerce_counts)
#df.groupby(by='district').agg({'industryField':lambda x:x.sum()},axis=0)
```

```python
df['salary1'] = df['salary'].agg(lambda x:x+500,axis=0)
df
```

```python
#计算各行政区的企业领域（industryField）包含电商的总数（通过匿名函数） --str.contains 返回识别字符串
df.groupby('district')['industryField'].apply(lambda x:x.str.contains('电商').sum())
```

```python
#transform 能够把整个DataFrame 的所有元素进行操作 --> 且能够返回原始 DataFrame 相同形状的结果
#在第1题数据的基础上，在原数据对象df 中新增一列，数值为该区的平均薪资水平   round() 表示保留x位小数
df['平均薪资水平'] = df.groupby('district')['salary'].transform(numpy.mean).round(2)
df
```
