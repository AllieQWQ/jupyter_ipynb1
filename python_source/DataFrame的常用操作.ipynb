```python
import csv
import pandas 
# 查看 DataFrame 的属性, values 表示聚合数据 ,index 表示索引 ,dtype 表示 返回数据类型 ,columns 表示获取列名
'''
    size,ndim,shape（分别表示查看元素个数、维度、形状） T表示转置
    
'''
data1 = pandas.read_csv('C:\\Users\\26440\\Desktop\\jupyter_ipynb\\python_source\\data\\某地区房屋销售数据.csv',encoding='gbk')
print(data1)
print(data1.values)
print(data1.columns)
print(data1.index)
print(data1.dtypes)
```

```python
dict = {
    'Name' : ['关羽','刘备','张飞','曹操'],
    'Age' : [28,34,29,42],
    'Salary' : [5000,8000,4500,10000],
    'hobby' : ['python','java','python','JavaScript']
}
data2 = pandas.DataFrame(dict,index=['rank1','rank2','rank3','rank4'])
print(data2)
# print(data2.index)
# print(data2.columns)
# print(data2.size)
# print(data2.dtypes)
# print(data2.shape)

print('第三行数据为\n',data2[:][2:3])    #第三行数据
print('Age列数据为：\n',data2.loc[:,'Age'])
print('Age和Salary列数据为：\n',data2.loc[:,['Age','Salary']])
print('前3行数据为：\n',data2.iloc[0:3,0:4])
print('后2行数据为：\n',data2.iloc[2:4,0:4])
print('前3行Name|Age|Salary数据为：\n',data2.iloc[0:3,0:3])     #索引也可以用[xx:xx]区间来写
print('取得行1和3,列Age和Salary数据为：\n',data2.loc[['rank1','rank3'],['Age','Salary']])
print('取得hobby列值为python的1、2列数据:\n',data2.loc[['rank1','rank3'],['Name','Age']])
print(data2.iloc[[0],0:4])  #只能取得一个数据
```

```python
'''
    查看访问dataframe的数据
    方法1：dataframe的数据基本查看方式
    方法2：单列数据（属性名称、字典【建议用字典访问】）
'''
data1.地区邮编 #属性名称！！ 使用访问属性的方法访问数据不建议使用
data1['房屋类型']   #字典
#case2:某一列的某几行
data1['房屋类型'][0:5]  #访问该列的前5行
data1['房屋类型'][1:3]
#case3:访问多列数据
data1[['房屋价格','房屋类型']]
#case4:多列的某几行
data1[['房屋价格','房屋类型']][:3]
#case5:所有列的某几行
data1[:][:3]
data1.tail()
```

```python
'''
    loc、iloc 的访问方式
    loc():索引名称的切片方法
    iloc():行索引和列索引的位置(0/1/2...)
'''
#case1: 单列切片
data1.loc[:,'房屋出售时间']
data1.iloc[:,1] #所有行的第一列
#case2:多列切片
data1.loc[:,['房屋出售时间','地区邮编']]
data1.iloc[:,0:3]   #所有行的1-3列  loc()前后均为闭区间     iloc() 为左闭右开区间
```
