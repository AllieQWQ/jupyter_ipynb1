```python
import csv
import requests

url = 'https://www.toutiao.com/hot-event/hot-board/?origin=toutiao_pc&_signature=_02B4Z6wo00f01BVDjygAAIDByhh8ZkIJI3QVZ4uAAGJ1adDlSckpFImylvV55AAsIKv3fZm9R6t7cDsZVK44yLwFL65zuQ2LCMIZEL9jrtIhGlpuau9fZYTVGAR2VHJL3uaLSH43yK7Z5wIV65'
res = requests.get(url)
dict = res.json()

# 定义CSV文件的头部，这里假设我们想要保存'Title'和'Hot'字段
headers = ['Title']

# 获取数据列表
dataList = dict['data']

# 创建一个字典列表，每个字典包含我们想要保存的字段
rows = [{'Title': data['Title']} for data in dataList]

# 保存CSV
with open('toutiao_data.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=headers)
    writer.writeheader()
    writer.writerows(rows)
```
