```python
import csv
import requests

all_doctor = []
for page in range(1,6):
    # 构造新的URL
    page_url = f'https://www.jkwin.com.cn/yst_web/doctor/getDoctorList?areaId=42&depId=&hasDuty=&pageNo={page}&pageSize=10'
   
    res = requests.get(page_url)

    dic = res.json()

    staffList = dic['doctors']
    for i in staffList:
        name = i['STAFF_NAME']
        Type = i['STAFF_TYPE']
        department = i['DEP_NAME']
        try:
            opinging = i['EV_NUM']
        except KeyError:
            opinging = 100

        item = [name,Type,department,opinging]
        all_doctor.append(item)
        print(item)

with open('doctor_data.csv','w',encoding='utf-8-sig',newline='')as f:
    writer = csv.writer(f)
    # 写入标题行
    writer.writerow(['姓名', '职称', '科室', '评价数'])
    # 写入所有医生信息
    writer.writerows(all_doctor)
```
