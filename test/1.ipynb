```python
import requests
import re
import csv

def fetch_douban_top250(start_page, end_page):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
    }

    moive_list = []

    for page in range(start_page, end_page + 1):
        url = f'https://movie.douban.com/top250?start={(page - 1) * 25}&filter='
        res = requests.get(url, headers=headers)
        html = res.text

        pattern = re.compile(
            r'<div class="item">.*?'
            r'<div class="pic">.*?<em .*?>(.*?)</em>.*?</div>.*?'
            r'<div class="info">.*?'
            r'<div class="hd">.*?<a .*?>(.*?)</a>.*?</div>.*?'
            r'<div class="bd">.*?<p.*?>(.*?)</p>.*?'
            r'<div class="star">.*?<span class="rating_num" property="v:average">(.*?)</span>.*?'
            r'<span>(.*?)人评价</span>.*?</div>.*?'
            r'</div>.*?</div>', re.S)

        items = re.findall(pattern, html)
        for item in items:
            dic = {}
            title = item[1].strip()
            title_cn = re.sub(r'<.*?>', '', title.split('\xa0/\xa0')[0]).strip()
            title_en = re.sub(r'<.*?>', '', title.split('\xa0/\xa0')[1]).strip() if '\xa0/\xa0' in title else ''
            links = re.search(r'<a href="(.*?)">', title)
            links = links.group(1) if links else ''
            
            info = item[2].strip()
            director = re.search(r'导演: (.*?)<br>', info)
            director = re.sub(r'<.*?>', '', director.group(1)).strip() if director else ''
            
            act = re.search(r'主演: (.*?)<br>', info)
            act = re.sub(r'<.*?>', '', act.group(1)).strip() if act else ''
            
            other_info = re.search(r'(\d{4})/(.*?)/(.*?)<br>', info)
            if other_info:
                Release_year = other_info.group(1)
                nationality = re.sub(r'<.*?>', '', other_info.group(2)).strip()
                genre = re.sub(r'<.*?>', '', other_info.group(3)).strip()
            else:
                Release_year, nationality, genre = '', '', ''
            
            score = item[3].strip()
            num_score = item[4].strip().replace('人评价', '')

            dic['电影中文名'] = title_cn
            dic['电影英文名'] = title_en
            dic['电影详情页链接'] = links
            dic['导演'] = director
            dic['主演'] = act
            dic['上映年份'] = Release_year
            dic['国籍'] = nationality
            dic['类型'] = genre
            dic['评分'] = score
            dic['评分人数'] = num_score
            moive_list.append(dic)

        print(f'----------------------第{page}页爬取完成--------------------------------------')
    print('-----------------------爬虫结束-------------------------------')

    with open('豆瓣电影Top250.csv', 'w', encoding='utf-8-sig', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=('电影中文名', '电影英文名', '电影详情页链接', '导演', '主演', '上映年份', '国籍', '类型', '评分', '评分人数'))
        writer.writeheader()
        writer.writerows(moive_list)

fetch_douban_top250(1, 10)
```
