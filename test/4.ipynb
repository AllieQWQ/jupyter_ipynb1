```python
import csv
import re
import requests

url = 'https://movie.douban.com/top250'

headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0 ; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36 Edg/130.0.0.0'
}

# 生成包含所有页面的URLs
urls = [f'https://movie.douban.com/top250?start={i * 25}' for i in range(10)]

count = 1

# 正则表达式用于匹配电影信息
movie_pattern = re.compile(
    r'<li>.*?<span class="title">(.*?)</span>.*?<p class="">.*?<br>(.*?)&nbsp.*?<span class="rating_num" property="v:average">(.*?)</span>.*?<span>(\d*)人评价</span>.*?<p class="quote">.*?<span class="inq">(.*?)</span>.*?</li>',
    re.S
)

# 正则表达式用于解析国家、类型、上映时间、时长
info_pattern = re.compile(r'(.*?)/(.*?)/(.*?)/(.*?)$')

with open('./douban_data.csv', 'w', newline='', encoding='utf-8-sig') as f:
    f_csv = csv.writer(f)
    f_csv.writerow(['Rank', 'Title', 'Country', 'Genre', 'Release Time', 'Duration', 'Score', 'Review Count', 'Quote'])

    for url in urls:
        response = requests.get(url=url, headers=headers)
        response.encoding = response.apparent_encoding
        html = response.text
        movies = re.findall(movie_pattern, html)

        for movie in movies:
            rank = count  # 排名
            title = movie[0].strip()  # 电影名称
            info = movie[1].strip().replace('\n', '').replace(' ', '')  # 国家/类型/上映时间/
            score = movie[2].strip()  # 评分
            review = movie[3].strip()  # 评论人数
            quote = movie[4].strip()  # 简评

            info_details = re.search(info_pattern, info)
            if info_details:
                country, genre, release_time, duration = info_details.groups()
            else:
                country, genre, release_time, duration = '', '', '', ''

            item = [rank, title, country, genre, release_time, duration, score, review, quote]
            f_csv.writerow(item)

            print(rank, title, country, genre, release_time, duration, score, review, quote)
            count += 1
```
