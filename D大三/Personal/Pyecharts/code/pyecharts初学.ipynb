## 使用pyecharts创建图表

```python
from pyecharts.charts import Line,Bar #仅导入echarts中的线图(加 , 可直接添加图表类型[首字母需大写])

#创建柱状图对象
bar = Bar()
bar.add_xaxis(['衬衫','羊毛衫','雪纺衫','裤子','高跟鞋','袜子'])
bar.add_yaxis('商家A',[11,22,33,44,55,66])

#渲染图表
bar.render_notebook() #在jupter中直接显示
#bar.render('bar.html') #用网页形式打开
```

# 标题配置项

```python
# TitleOpts
from pyecharts import options as opts #导入配置项包


bar4 = (
    Bar()
    .add_xaxis(Faker.choose())
    .add_yaxis('图例A',Faker.values())
    .add_yaxis('图例B',Faker.values())
    .set_global_opts(
        title_opts = opts.TitleOpts(title='这个是配置项的标题',
                                    subtitle='这个是副标题',pos_left='center'), #使用标题配置项进行实例化
        legend_opts = opts.LegendOpts(pos_right='10%',pos_bottom='90%',orient='vertical',
                                    border_color='red',background_color='yellow',border_width=1), #图例配置项，同样需要实例化 （本质上【left和right】【top和bottom】的功能相似）
        tooltip_opts = opts.TooltipOpts(trigger='item'),
    )
)

# 渲染图表
#bar4.render('bar3.html')
bar4.render_notebook()
```

# 缩放条、坐标轴 配置项（DataZoomOpts、AxisOpts）

- 使用与坐标轴相关大全局配置项大时候，调用要在前面添加x 和 y

```python
line = (
    Line()   #初始化配置项
    .add_xaxis(Faker.choose())
    .add_yaxis('图例A',Faker.values())
    .add_yaxis('图例B',Faker.values())
    .set_global_opts(
        title_opts = opts.TitleOpts(title='这个是配置项的标题',
                                    subtitle='这个是副标题',pos_left='center'), #使用标题配置项进行实例化
        legend_opts = opts.LegendOpts(pos_right='10%',pos_bottom='90%',orient='vertical',
                                    border_color='red',border_width=1), #图例配置项，同样需要实例化 （本质上【left和right】【top和bottom】的功能相似）
        tooltip_opts = opts.TooltipOpts(trigger='axis',background_color='yellow',axis_pointer_type='shadow'),
        datazoom_opts = opts.DataZoomOpts(type_='slider'), #type_选择slider可自己滑动、inside滚轮控制
        yaxis_opts = opts.AxisOpts(max_=200), #给y轴设置上限200
        xaxis_opts = opts.AxisOpts(is_show= False),
    )
)

# 渲染图表
line.render('line1.html')
#line.render_notebook()
```

## 初始化配置项
- 可以设置画布长宽、图表ID、标题、图表背景色等

```python
from pyecharts.charts import Line,Bar
from pyecharts.faker import Faker
from pyecharts import options as opts #导入配置项包
from pyecharts.globals import ThemeType #导入主题配置项(这个可以进行定制主题)
line2 = (
    Line(init_opts = opts.InitOpts(width="1200px",height="800px",theme=ThemeType.DARK,page_title="折线图"))   #初始化配置项(主题颜色会和背景颜色冲突)
    .add_xaxis(Faker.choose())
    .add_yaxis('图例A',Faker.values())
    .add_yaxis('图例B',Faker.values())
    .set_global_opts(
        title_opts = opts.TitleOpts(title='这个是配置项的标题',
                                    subtitle='这个是副标题',pos_left='center'), #使用标题配置项进行实例化
        legend_opts = opts.LegendOpts(pos_right='10%',pos_bottom='95%',orient='vertical',
                                    border_color='red',border_width=1), #图例配置项，同样需要实例化 （本质上【left和right】【top和bottom】的功能相似）
        tooltip_opts = opts.TooltipOpts(trigger='axis',background_color='yellow',axis_pointer_type='shadow'),
               
    )
)

# 渲染图表
line2.render('../HTML/line2.html')
#line2.render_notebook()
```

## 系列配置项
- 直接像全局配置项一样去使用
### 标签配置项（LabelOpts）

```python
line3 = (
    Line(init_opts = opts.InitOpts(width="1200px",height="800px",theme=ThemeType.DARK,page_title="折线图"))   #初始化配置项(主题颜色会和背景颜色冲突)
    .add_xaxis(Faker.choose())
    .add_yaxis('图例A',Faker.values())
    .add_yaxis('图例B',Faker.values())
    .set_global_opts(
        title_opts = opts.TitleOpts(title='这个是配置项的标题',
                                    subtitle='这个是副标题',pos_left='center'), #使用标题配置项进行实例化
        legend_opts = opts.LegendOpts(pos_right='10%',pos_bottom='95%',orient='vertical',
                                    border_color='red',border_width=1), #图例配置项，同样需要实例化 （本质上【left和right】【top和bottom】的功能相似）
        tooltip_opts = opts.TooltipOpts(trigger='axis',background_color='yellow',axis_pointer_type='shadow'),               
    )
    .set_series_opts(  # 系列配置项 就是用series 进行控制  is_show=True(默认为显示)
        label_opts = opts.LabelOpts(position="right",color="orange",formatter='{a}-{c}',font_family='Times New Roman'),   #formatter 给予他们数值名称、数据名称（{a},{b},{c}...）
        markpoint_opts = opts.MarkPointOpts(
            data = [opts.MarkPointItem(type_='max',name="最大值"),
             opts.MarkPointItem(type_='min',name="最小值")]
        ),
    )
)

# 渲染图表
line3.render('line3.html')
#line3.render_notebook()
```

### 标签点、线、区域配置项（MarkPointsOpts、MarkLineOpts、MarkAreaOpts）

```python
line4 = (
    Line(opts.InitOpts(theme=ThemeType.DARK,page_title="折线图"))   #初始化配置项(主题颜色会和背景颜色冲突)
    .add_xaxis(Faker.choose())
    .add_yaxis('图例A',Faker.values())
    .add_yaxis('图例B',Faker.values())
    .set_global_opts(
        title_opts = opts.TitleOpts(title='这个是配置项的标题',
                                    subtitle='这个是副标题',pos_left='center'), #使用标题配置项进行实例化
        legend_opts = opts.LegendOpts(pos_right='10%',pos_bottom='95%',orient='vertical',
                                    border_color='red',border_width=1), #图例配置项，同样需要实例化 （本质上【left和right】【top和bottom】的功能相似）
        tooltip_opts = opts.TooltipOpts(trigger='axis',background_color='yellow',axis_pointer_type='shadow'),               
    )
    .set_series_opts(  # 系列配置项 就是用series 进行控制  is_show=True(默认为显示)
        label_opts = opts.LabelOpts(position="right",color="orange",formatter='{c}',font_family='方正小标宋_GBK'), #formatter 给予他们数值名称、数据名称（{a},{b},{c}...）
        markpoint_opts = opts.MarkPointOpts(
            data = [
                opts.MarkPointItem(type_='max'),
                opts.MarkPointItem(type_='min'),
                opts.MarkPointItem(type_='average'),#这里是最大值和最小值标记点的配置项及参数
            ]
        ),
        
        markline_opts = opts.MarkLineOpts(
            data = [
                opts.MarkLineItem(type_='max',name='这是最大值'),
                opts.MarkLineItem(type_='min',name='这是最小值'),
                opts.MarkLineItem(type_='average',name='这是平均值'),  #这里是最大值和最小值标记线的配置项及参数
            ]
        )
        
    )
)

# 渲染图表
line4.render('line4.html')
#line4.render_notebook()
```

```python
line4 = (
    Line(opts.InitOpts(theme=ThemeType.DARK,page_title="标记点、线配置项"))   #初始化配置项(主题颜色会和背景颜色冲突)
    .add_xaxis(Faker.choose())
    .add_yaxis('图例A',Faker.values(),
              markpoint_opts = opts.MarkPointOpts(
            data = [
                opts.MarkPointItem(type_='max',symbol_size=[40,40],symbol='diamond'),
                opts.MarkPointItem(type_='min'),
                opts.MarkPointItem(type_='average'),#这里是最大值和最小值标记点的配置项及参数
            ]
        ))
    .add_yaxis('图例B',Faker.values(),
              markline_opts = opts.MarkLineOpts(
            data = [
                opts.MarkLineItem(type_='max',name='这是最大值'),
                opts.MarkLineItem(type_='min',name='这是最小值'),
                opts.MarkLineItem(type_='average',name='这是平均值'),  #这里是最大值和最小值标记线的配置项及参数
            ]
        )
              )
    .set_global_opts(
        title_opts = opts.TitleOpts(title='这个是配置项的标题',
                                    subtitle='这个是副标题',pos_left='center'), #使用标题配置项进行实例化
        legend_opts = opts.LegendOpts(pos_right='10%',pos_bottom='95%',orient='vertical',
                                    border_color='red',border_width=1), #图例配置项，同样需要实例化 （本质上【left和right】【top和bottom】的功能相似）
        tooltip_opts = opts.TooltipOpts(trigger='axis',background_color='yellow',axis_pointer_type='shadow'),               
    )
    .set_series_opts(  # 系列配置项 就是用series 进行控制  is_show=True(默认为显示)
        label_opts = opts.LabelOpts(position="right",color="orange",formatter='{c}',font_family='方正小标宋_GBK'), #formatter 给予他们数值名称、数据名称（{a},{b},{c}...）          
    )
)

# 渲染图表
line4.render('line4.html')
#line4.render_notebook()
```

### 图元样式配置项

```python
line5 = (
    Line(opts.InitOpts(page_title="图元配置项"))   #初始化配置项(主题颜色会和背景颜色冲突)
    .add_xaxis(Faker.choose())
    .add_yaxis('图例A',Faker.values(),
               emphasis_opts = opts.EmphasisOpts(
                    itemstyle_opts=opts.ItemStyleOpts(color='yellow'),
                    linestyle_opts=opts.LineStyleOpts(width=7),
                    areastyle_opts=opts.AreaStyleOpts(opacity=0.3,color="yellow")
                )
               
              ) #border_type='dashed'仅能用于柱状图和饼图等
    .add_yaxis('图例B',Faker.values(),
               emphasis_opts = opts.EmphasisOpts(
                   itemstyle_opts=opts.ItemStyleOpts(color='red'),
                   linestyle_opts=opts.LineStyleOpts(type_="dashed",width=5),
                   areastyle_opts=opts.AreaStyleOpts(opacity=0.3,color="red")
               )              
              )
    .set_global_opts(
        title_opts = opts.TitleOpts(title='这个是配置项的标题',
                                    subtitle='这个是副标题',pos_left='center'), #使用标题配置项进行实例化
        legend_opts = opts.LegendOpts(pos_right='10%',pos_bottom='95%',orient='vertical',
                                    border_color='red',border_width=1), #图例配置项，同样需要实例化 （本质上【left和right】【top和bottom】的功能相似）
        tooltip_opts = opts.TooltipOpts(trigger='axis',background_color='yellow',axis_pointer_type='shadow',),   
        
    )
    .set_series_opts(  # 系列配置项 就是用series 进行控制  is_show=True(默认为显示)
        label_opts = opts.LabelOpts(position="right",color="blue",formatter='{c}',font_family='方正小标宋_GBK'), #formatter 给予他们数值名称、数据名称（{a},{b},{c}...） 
        #itemstyle_opts = opts.ItemStyleOpts(color='red'),  #这个是系列配置项里的图文样式（但是很单一）
        emphasis_opts = opts.EmphasisOpts(is_scale=True,focus='series'),   #突显样式
        )
)

# 渲染图表
line5.render('../HTML/line5.html')
#line5.render_notebook()
```
